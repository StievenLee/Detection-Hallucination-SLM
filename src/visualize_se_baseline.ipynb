```python
# ============================================================
# Hitung F1-SQuAD (rata-rata per model x dataset) dari se_results.csv
# File summary (se_auroc_summary.csv) TIDAK menyimpan f1_squad,
# sehingga kita ambil dari se_results.csv (per-pertanyaan) lalu rata-ratakan.
# ============================================================

FACTOID_DATASETS = {'trivia_qa', 'bioasq', 'facqa'}  # WReTE pakai macro-F1, bukan SQuAD-F1

def mean_f1_squad_per_dataset(results_path):
    """Baca se_results.csv secara robust dan kembalikan dict {dataset: mean_f1_squad}.
    Hanya untuk dataset factoid; baris dengan f1_squad kosong/rusak diabaikan."""
    # engine='python' + on_bad_lines='skip' agar tahan terhadap CSV yang quoting-nya cacat
    df = pd.read_csv(
        results_path,
        engine='python',
        on_bad_lines='skip',
        dtype=str,
        quotechar='"',
    )
    df.columns = [c.strip() for c in df.columns]
    if 'dataset' not in df.columns or 'f1_squad' not in df.columns:
        print(f'  [warn] kolom dataset/f1_squad tidak ditemukan di {results_path}')
        return {}
    df['dataset'] = df['dataset'].astype(str).str.strip()
    df['f1_squad'] = pd.to_numeric(df['f1_squad'], errors='coerce')
    out = {}
    for ds, grp in df.groupby('dataset'):
        if ds not in FACTOID_DATASETS:
            continue
        valid = grp['f1_squad'].dropna()
        if len(valid) > 0:
            out[ds] = valid.mean()
    return out
```

```python
# ============================================================
# Gabungkan kolom f1_squad ke setiap tabel summary
# (mendukung >1 file se_results per model, mis. Phi yang terpecah dua batch)
# ============================================================

# Bisa berupa 1 path (str) atau beberapa path (list) per model
RESULTS_PATHS = {
    'tinyllama': ['../results/metrics/tinyllama/se_results.csv'],
    'phi':       ['../results/metrics/phi/se_results.csv',
                  '../results/metrics/phi/se_results(2).csv'],  # batch kedua Phi (jika ada)
    'qwen':      ['../results/metrics/qwen/se_results.csv'],
}

def mean_f1_squad_multi(paths):
    """Gabungkan beberapa se_results.csv lalu hitung mean f1_squad per dataset factoid."""
    if isinstance(paths, str):
        paths = [paths]
    merged = {}
    for p in paths:
        try:
            m = mean_f1_squad_per_dataset(p)
        except FileNotFoundError:
            print(f'  [info] lewati (tidak ada): {p}')
            continue
        # jika dataset muncul di lebih dari satu file, ambil yang pertama non-kosong
        for ds, val in m.items():
            merged.setdefault(ds, val)
    return merged

def attach_f1_squad(summary_df, paths):
    summary_df = summary_df.copy()
    summary_df['dataset'] = summary_df['dataset'].astype(str).str.strip()
    f1_map = mean_f1_squad_multi(paths)
    if 'f1_squad' not in summary_df.columns:
        summary_df['f1_squad'] = np.nan
    for i, row in summary_df.iterrows():
        ds = row['dataset']
        if ds in f1_map:
            summary_df.at[i, 'f1_squad'] = round(f1_map[ds], 4)
    return summary_df

result_tinyllama = attach_f1_squad(result_tinyllama, RESULTS_PATHS['tinyllama'])
result_phi       = attach_f1_squad(result_phi,       RESULTS_PATHS['phi'])
try:
    result_phi2  = attach_f1_squad(result_phi2,      RESULTS_PATHS['phi'])
except NameError:
    pass
result_qwen      = attach_f1_squad(result_qwen,      RESULTS_PATHS['qwen'])

cols_show = ['model', 'dataset', 'language', 'accuracy', 'auroc', 'f1_squad', 'wrete_macro_f1']
def show(df):
    existing = [c for c in cols_show if c in df.columns]
    return df[existing]

show(result_tinyllama)
```

```python
# ============================================================
# Tabel F1 final: satukan ketiga model, pilih F1 sesuai jenis tugas
#   - factoid (trivia_qa, bioasq, facqa) -> f1_squad
#   - wrete                              -> wrete_macro_f1
# ============================================================

def build_f1_table(*dfs):
    frames = []
    for df in dfs:
        if df is None:
            continue
        d = df.copy()
        d['dataset'] = d['dataset'].astype(str).str.strip()
        # tentukan kolom F1 final per baris
        def pick_f1(r):
            ds = r['dataset']
            if ds == 'wrete':
                return r.get('wrete_macro_f1', np.nan)
            return r.get('f1_squad', np.nan)
        d['f1'] = d.apply(pick_f1, axis=1)
        d['f1_metric'] = np.where(d['dataset'].eq('wrete'), 'macro-F1', 'SQuAD-F1')
        frames.append(d[['model', 'dataset', 'f1_metric', 'f1', 'accuracy', 'auroc']])
    if not frames:
        return pd.DataFrame()
    out = pd.concat(frames, ignore_index=True)
    # urutkan dataset supaya rapi
    order = {'trivia_qa': 0, 'bioasq': 1, 'facqa': 2, 'wrete': 3}
    out['_o'] = out['dataset'].map(order).fillna(9)
    out = out.sort_values(['_o', 'model']).drop(columns='_o').reset_index(drop=True)
    return out

f1_table = build_f1_table(
    result_tinyllama,
    result_phi,
    result_phi2 if 'result_phi2' in dir() else None,
    result_qwen,
)
f1_table
```

## Catatan
- File se_auroc_summary.csv tidak menyimpan f1_squad; nilai ini dihitung dari se_results.csv (per-pertanyaan) lalu dirata-ratakan.
- Dataset factoid (TriviaQA, BioASQ, FacQA) memakai SQuAD token-level F1; WReTE memakai macro-F1 (wrete_macro_f1).
- pivot_f1 di bawah bisa langsung disalin ke Tabel F1 di paper.

```python
# Pivot: baris = dataset, kolom = model, nilai = F1 (siap salin ke Tabel F1 paper)
def short_model(m):
    m = str(m).lower()
    if 'tinyllama' in m: return 'TinyLlama'
    if 'phi'       in m: return 'Phi-1.5'
    if 'qwen'      in m: return 'Qwen1.5'
    return m

if not f1_table.empty:
    t = f1_table.copy()
    t['model'] = t['model'].map(short_model)
    pivot_f1 = t.pivot_table(index=['dataset', 'f1_metric'],
                             columns='model', values='f1', aggfunc='first')
    display(pivot_f1)
else:
    print('f1_table kosong.')
```
